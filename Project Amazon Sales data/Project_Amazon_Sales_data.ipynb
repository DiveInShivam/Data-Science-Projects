**3. Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sales distribution by region

plt.figure(figsize=(10, 6))
sns.countplot(y='Region', data=data, order=data['Region'].value_counts().index)
plt.title('Sales Distribution by Region')
plt.xlabel('Number of Sales')
plt.ylabel('Region')
plt.show()
```

```python
# Total Revenue by Item Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Revenue', y='Item Type', data=data, estimator=sum, errorbar=None, order=data.groupby('Item Type')['Total Revenue'].sum().sort_values().index)
plt.title('Total Revenue by Item Type')
plt.xlabel('Total Revenue')
plt.ylabel('Item Type')
plt.show()
```

```python
# Online vs Offline Sales
plt.figure(figsize=(12, 6))
sns.countplot(x='Sales Channel', data=data)
plt.title('Online vs Offline Sales')
plt.xlabel('Sales Channel')
plt.ylabel('Number of Sales')
plt.show()
```

**4. Detailed data analysis**

```python
# Sales Trends Over Time
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], errors='coerce')
#  sum of 'Total Revenue' for each month
sales_trend = data.groupby(data['Order Date'].dt.to_period('M'))['Total Revenue'].sum().reset_index()
sales_trend['Order Date'] = sales_trend['Order Date'].dt.to_timestamp()
plt.figure(figsize=(12, 6))
plt.plot(sales_trend['Order Date'], sales_trend['Total Revenue'], marker='o')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Profitability Analysis by Country and Region
profit_by_country = data.groupby(['Region', 'Country'])['Total Profit'].sum().reset_index().sort_values(by='Total Profit', ascending=False)

plt.figure(figsize=(12, 12))
sns.barplot(x='Total Profit', y='Country', data=profit_by_country.head(20), hue='Region')
plt.title('Top 20 Countries by Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('Country')
plt.legend(title='Region', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Customer Behavior Analysis: Sales Channel and Order Priority
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sales Channel', y='Total Profit', data=data)
plt.title('Total Profit by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Profit')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Order Priority', y='Total Profit', data=data)
plt.title('Total Profit by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Product Performance Analysis
product_performance = data.groupby('Item Type')[['Total Revenue', 'Total Cost', 'Total Profit']].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Total Revenue', y='Item Type', data=product_performance.sort_values(by='Total Revenue', ascending=False))
plt.title('Total Revenue by Item Type')
plt.xlabel('Total Revenue')
plt.ylabel('Item Type')
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
